```python
df = pd.read_csv("query.csv")

color_scale=["#1a9850", "#91cf60", "#d9ef8b", "#ffffbf", "#fee08b", "#fc8d59", "#d73027"]

m = folium.Map(location=[55.404, 9.531],
               zoom_start=4, 
               min_zoom=4,
               tiles="https://{s}.tile.opentopomap.org/{z}/{x}/{y}.png", 
               attr='Map data: &copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors, <a href="http://viewfinderpanoramas.org">SRTM</a> | Map style: &copy; <a href="https://opentopomap.org">OpenTopoMap</a> (<a href="https://creativecommons.org/licenses/by-sa/3.0/">CC-BY-SA</a>)', 
               prefer_canvas=True, 
               overlay=True)

for lat, lon, mag, place in zip(df['latitude'], df['longitude'], df['mag'], df['place']):
  folium.CircleMarker(location=[lat, lon], radius=mag*3, color="#000000", weight=3, fill_color=color_scale[int(mag//1)], fill_opacity=0.6, popup=folium.Popup("<b>Location:</b> {0}<br><b>Magnitude:</b> {1}".format(place, mag), width=500)).add_to(m)

m
```
